```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
gas = pd.read_csv('gas_prices.csv')
gas
```

```python
plt.figure(figsize=(10, 6))
plt.title('Gas Price US$')

list_country = gas.columns.tolist()

for country in gas.columns[1:]:  # Assume-se que a primeira coluna é 'Year'
    if country in list_country:
        plt.plot(gas['Year'], gas[country], label=country)

plt.xlabel('Year')
plt.ylabel('Price')
plt.xticks(gas['Year'][::2])  # Exibir rótulos do eixo x a cada 2 anos

plt.legend()
plt.savefig('Gas_price.png')
plt.show()

```
