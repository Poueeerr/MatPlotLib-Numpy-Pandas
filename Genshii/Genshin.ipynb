```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as mpl
import re
```

```python
dados.groupby(['region']).count().sort_values('character_name', ascending = False)
```

```python
region_counts = dados['region'].value_counts()

mpl.figure(figsize=(20, 6))
mpl.subplot(131)
mpl.bar(region_counts.index, region_counts.values)
mpl.xlabel('Region', labelpad=10)
mpl.ylabel('Count of Character Names')
mpl.title('Character Name Counts by Region')
mpl.tight_layout()
```
